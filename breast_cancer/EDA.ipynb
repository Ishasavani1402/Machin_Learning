```python
import math
import matplotlib.pyplot as plt

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

n_cols = 2
n_rows = math.ceil(len(numeric_cols) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    axes[i].hist(df[col].dropna(), bins=30)
    axes[i].set_title(col)

# remove extra empty plots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# distribution plot for numeric columns by exit columns
import seaborn as sns
numeric_cols = ['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean']
for col in numeric_cols:
    plt.figure(figsize=(8,4))
    for dig_status in df["diagnosis"].unique():
        sns.histplot(df[df["diagnosis"] == dig_status][col], kde=True, label=f"Exited: {dig_status}")
    plt.title(f"Distribution of {col} by dig Status")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()
```
